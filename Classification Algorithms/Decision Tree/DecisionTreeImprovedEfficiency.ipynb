```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("loan_prediction.csv")
data
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data["LoanAmount"] = data["LoanAmount"].fillna(data["LoanAmount"].mean())
data["LoanAmount"].isnull().sum()
```

```python
data["Loan_Amount_Term"] = data["Loan_Amount_Term"].fillna(data["Loan_Amount_Term"].mean())
data["Loan_Amount_Term"].isnull().sum()
```

```python
data["Credit_History"] = data["Credit_History"].fillna(data["Credit_History"].mean())
data["Credit_History"].isnull().sum()
```

```python
data["Gender"] = data["Gender"].fillna(method = "ffill")
data["Gender"].isnull().sum()
```

```python
data["Married"] = data["Married"].fillna(method = "ffill")
data["Dependents"] = data["Dependents"].fillna(method = "ffill")
data["Self_Employed"] = data["Self_Employed"].fillna(method = "ffill")
```

```python
data.isnull().sum()
```

```python
data = data.drop(columns = "Loan_ID")
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
data["Gender"] = le.fit_transform(data["Gender"])
data["Married"] = le.fit_transform(data["Married"])
data["Education"] = le.fit_transform(data["Education"])
data["Self_Employed"] = le.fit_transform(data["Self_Employed"])
data["Loan_Status"] = le.fit_transform(data["Loan_Status"])
data["Dependents"] = le.fit_transform(data["Dependents"])
data["Property_Area"] = le.fit_transform(data["Property_Area"])
```

```python
data.info()
```

```python
x = data.drop(columns = "Loan_Status")
x.head()
```

```python
y = data.Loan_Status
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
```
